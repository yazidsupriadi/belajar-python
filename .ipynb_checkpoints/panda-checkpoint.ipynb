### Persiapan membuat Data Frame

untuk membuat dataframe dapat ditulisakn dengan contoh sintaks sebagai berikut:

pandas.DataFrame( data, index, columns, dtype, copy)

1. data = merupakan data yang ingin dipakai.
2. index = merupakan label untuk baris.
3. columns - merupakan label untuk kolom.
4. dtypes = merupakan tipe data yang digunakan.
5. copy = digunakan untuk menyalin data , secara default bernilai false.

```python
jumlah_baris = 10
jumlah_kolom = 5
cols = tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1,5,size = (jumlah_baris,jumlah_kolom)),columns =cols )
print(df)
```

##### menambahkan prefix

```python
#menambahkan prefix
df.add_prefix('Kolom_')
```

```python
#menambahkan suffix
df.add_suffix(' Plot')
```

#####  menyeleksi dengan operator logika |

```python
#selection with |
df[(df['A'] == 2) | (df['A'] == 3)]
```

```python
#selection with isin()
df[(df['A'].isin([2,3]))]
```

##### Menyeleksi dengan negasi ~ 

```python
#selection with negation ~
df[~(df['A'].isin([2,3]))]
```

### Sorting dan Describing Data

```python
#describe data
df.describe()
```

```python
#sorting data
df.sort_values(['Name'])
#secara ascending / descending
df.sort_values('Name',ascending = 'False')
#sorting dengan 2 kolom
df.sort_values(['Name','Type 2'],ascending ='True')
#sorting dengan ascending/descending tertentu
df.sort_values(['Name','Type 2'],ascending =[0,1])
```
